```python
from typing import List

class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        window_sum = sum(nums[:k])
        max_sum = window_sum

        for index in range(k, len(nums)):
            ele_to_sub = nums[index - k]
            ele_to_add = nums[index]

            window_sum += ele_to_add - ele_to_sub

            if window_sum > max_sum:
                max_sum = window_sum

        return max_sum/k
```

```python
from typing import List

class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        window_sum = sum(nums[:k])
        max_sum = window_sum

        for index in range(k, len(nums)):
            window_sum += nums[index] - nums[index - k]

            if window_sum > max_sum:
                max_sum = window_sum

        return max_sum/k
```

```python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        window_set = set()

        for i in range(len(nums)):
            if nums[i] in window_set:
                return True

            window_set.add(nums[i])

            if len(window_set) > k:
                window_set.remove(nums[i-k])
        
        return False
```

```python
class Solution:
    def numOfSubarrays(self, arr: List[int], k: int, threshold: int) -> int:
        target_sum = k * threshold

        window_sum = sum(arr[:k])
        count = 1 if window_sum >= target_sum else 0

        for i in range(k, len(arr)):
            window_sum += arr[i] - arr[i-k]
            if window_sum >= target_sum:
                count += 1
    
        return count
```

```python
class Solution:
    def decrypt(self, code: List[int], k: int) -> List[int]:
        n = len(code)
        res = [0] * n
        window_sum = 0
        
        if k > 0:
            for t in range(1, k + 1):
                window_sum += code[t % n]

            for i in range(n):
                res[i] = window_sum
                window_sum += code[(i + k + 1) % n] - code[(i + 1) % n]
        
        elif k < 0:
            k = -k
            for t in range(1, k + 1):
                t = n - t
                window_sum += code[t % n]

            for i in range(n):
                res[i] = window_sum
                window_sum += code[i % n] - code[(i - k) % n]
  
        return res
```

```python
from typing import List

class Solution:
    def decrypt(self, code: List[int], k: int) -> List[int]:
        n = len(code)
        res = [0] * n

        if k == 0:
            return res

        neg = False
        if k < 0:
            neg = True
            k = -k
            code.reverse() 

        window_sum = 0
        for t in range(1, k + 1):
            window_sum += code[t % n]

        res[0] = window_sum

        for i in range(1, n):
            window_sum += code[(i + k) % n] - code[i]
            res[i] = window_sum

        if neg:
            res.reverse()

        return res
```

```python
from typing import List

class Solution:
    def dietPlanPerformance(self, calories: List[int], k: int, lower: int, upper: int) -> int:
        n = len(calories)
        points = 0

        window_sum = sum(calories[:k])

        if window_sum < lower:
            points -= 1
        elif window_sum > upper:
            points += 1

        for i in range(k, n):
            window_sum += calories[i] - calories[i - k]

            if window_sum < lower:
                points -= 1
            elif window_sum > upper:
                points += 1

        return points
```
